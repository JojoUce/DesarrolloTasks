```python
# Importar dependencias necesarias para el análisis de datos y procesamiento de HTML
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
# Función que convierte una lista de elementos en un DataFrame de Pandas
def rowsToDataFrame(rows):
    # Si el número de columnas no coincide, ajustamos el número de columnas al máximo encontrado
    max_cols = max(len(row) for row in rows)
    df = pd.DataFrame(rows[1:], columns=rows[0] + [''] * (max_cols - len(rows[0])))
    return df
```

```python
# Función para extraer los datos de una tabla HTML y convertirla en una lista de listas
def processTableData(tbl):
    rows = []
    for child in tbl.find('tbody').children:
        row = []
        for td in child:
            try:
                item = td.text.replace('\n', '')
                if item:
                    row.append(item)
            except:
                continue
        if len(row) > 0:
            rows.append(row)
    return rows
```

```python
# Leer contenido de la web con Requests
r = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)')
```

```python
# Procesar el contenido HTML recuperado y convertirlo en una tabla de datos
soup = bs(r.content, 'html.parser')
tbl = soup.find_all('table', {'class': 'wikitable'})[0]
```

```python
# Procesar la tabla y convertirla en una lista de listas
table = processTableData(tbl)
```

```python
# Convertir la lista de filas de la tabla en un DataFrame de Pandas
df = rowsToDataFrame(table)
```

```python
# Guardamos los datos extraídos en un archivo CSV para usarlos posteriormente
df.to_csv('countries_by_GDP.csv', index=False)
```
