```python
verbos = {}
hashtags = []

for texto in textos:
    # Analizar el texto utilizando spaCy
    doc = nlp(texto)

    # Contar los verbos en el texto
    for token in doc:
        if token.pos_ == 'VERB':
            verbo = token.lemma_
            verbos[verbo] = verbos.get(verbo, 0) + 1

    # Encontrar hashtags en el texto
    hashtags.extend(re.findall(r'#\w+', texto))
```

```python
# Imprimir los resultados
for verbo, cantidad in verbos.items():
    print(f"{verbo} = {cantidad}")
```

```python
print("\nHashtags encontrados:")
for hashtag in hashtags:
    print(hashtag)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(verbos.keys(), verbos.values(), color='skyblue')
plt.xlabel("Verbos")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de verbos")
plt.xticks(rotation=45)
plt.show()
```

```python
hashtags_count = {hashtag: hashtags.count(hashtag) for hashtag in set(hashtags)}
```
